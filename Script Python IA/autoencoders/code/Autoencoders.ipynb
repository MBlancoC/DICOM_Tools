```python
def cross_validation_stratified(clf, X_np, y_np, cv):
  skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=1)
  lst_accu_stratified = []
  
  X = X_np.copy()
  X['y'] = y_np
  y = X['y']
  X = X.drop(columns=['y'])

  for train_index, test_index in skf.split(X, y):
      x_train_fold, x_test_fold = X.iloc[train_index], X.iloc[test_index]
      y_train_fold, y_test_fold = y[train_index], y[test_index]
      clf.fit(x_train_fold, y_train_fold)
      lst_accu_stratified.append(clf.score(x_test_fold, y_test_fold))
  return np.asarray(lst_accu_stratified) 
```

```python
class encoder_clf():
    def __init__(self, autoencoder):
        self.autoencoder_obj = autoencoder
        self.model_with_clf  = None
        self.model_without_clf = None        

    def fit(self, num_clases, x_train, y_train, x_test, y_test, epochs, batch_size, shuffle, check=False):
      self.model_with_clf = self.add_clasificador(num_clases, x_train, y_train, x_test, y_test, epochs, batch_size, shuffle)
      self.model_without_clf = self.remove_clasificador()
      if check:
        self.check()
      return self.model_with_clf

    def clone_model(self, model):
      clone =  keras.models.clone_model(model)
      clone.set_weights(model.get_weights())
      return clone

    def compile_and_fit(self, model, x_train, y_train, x_test, y_test, epochs, batch_size, shuffle):
      model.compile(loss='categorical_crossentropy', optimizer=self.autoencoder_obj.optimizer)

      history = model.fit(x_train, y_train,
                  epochs=epochs,
                  batch_size=batch_size,
                  shuffle=shuffle,
                  callbacks=[CALLBACK_ES],
                  verbose = self.autoencoder_obj.verbose,
                  validation_data=(x_test, y_test))  
      
      return model,  history

    def check(self):
      print('----  check  ---')
      self.model_with_clf.summary()
      self.model_without_clf.summary()
      print(' Encoder original vs Encoder Frezze + softmax -> TRUE')
      for i in range(0, len(self.model_without_clf.get_weights())):
          print('-- capa:', i)
          print(self.model_with_clf.get_weights()[i]==self.model_without_clf.get_weights()[i])
      

    def add_clasificador(self, num_clases, x_train, y_train, x_test, y_test, epochs, batch_size, shuffle):
 
      input_layer = Input(shape=(self.autoencoder_obj.input_dim,))
      cont = 0
      encoding_layers = []
      for i in range(len(self.autoencoder_obj.capas)):
         if i == 0:
              encoding_layers.append(Dense(self.autoencoder_obj.capas[i], activation='relu')(input_layer))
         else:
              encoding_layers.append(Dense(self.autoencoder_obj.capas[i], activation='relu')(encoding_layers[i-1]))
      #latent_layer = Dense(self.encoding_dim, activation='relu')(encoding_layers[-1])
      if len(encoding_layers):
        latent_layer = Dense(self.autoencoder_obj.encoding_dim, activation='sigmoid')(encoding_layers[-1])
      else:
        latent_layer = Dense(self.autoencoder_obj.encoding_dim, activation='sigmoid')(input_layer)

      latent_layer = Dense(self.autoencoder_obj.encoding_dim, activation='sigmoid')(input_layer)
      clf_layer = Dense(num_clases, activation='softmax')(latent_layer)

      new_model = Model(inputs=input_layer, outputs=clf_layer)            

      new_model, history = self.compile_and_fit(new_model, x_train, y_train, x_test, y_test, epochs, batch_size, shuffle)
      return new_model

    def remove_clasificador(self):

      input_layer = Input(shape=(self.autoencoder_obj.input_dim,))
      cont = 0
      encoding_layers = []
      for i in range(len(self.autoencoder_obj.capas)):
         if i == 0:
              encoding_layers.append(Dense(self.autoencoder_obj.capas[i], activation='relu')(input_layer))
         else:
              encoding_layers.append(Dense(self.autoencoder_obj.capas[i], activation='relu')(encoding_layers[i-1]))
      #latent_layer = Dense(self.encoding_dim, activation='relu')(encoding_layers[-1])
      if len(encoding_layers):
        latent_layer = Dense(self.autoencoder_obj.encoding_dim, activation='sigmoid')(encoding_layers[-1])
      else:
        latent_layer = Dense(self.autoencoder_obj.encoding_dim, activation='sigmoid')(input_layer)

      latent_layer = Dense(self.autoencoder_obj.encoding_dim, activation='sigmoid')(input_layer)      
      
      new_model = Model(inputs=input_layer, outputs=latent_layer)        
      #copiamos pesos del encoder con clasificador
      for i in range(0, len(self.model_with_clf.layers)-1):
        #print(i)
        new_model.layers[i].set_weights(self.model_with_clf.layers[i].get_weights())
      return new_model      


    def save(self, nameFile):
      self.autoencoder_obj.save(nameFile+'_encoder_autoencoder_obj_clf.h5')
      self.model_with_clf.save(nameFile+'_encoder_clf_model_with_clf.h5')
      self.model_without_clf.save(nameFile+'_encoder_clf_model_without_clf.h5')

    def load(self, nameFile):
      self.autoencoder_obj = myAutoencoder(encoding_dim=None, input_dim=None, loss=None, optimizer=None)
      self.autoencoder_obj.load(nameFile + '_encoder_autoencoder_obj_clf.h5')
      self.model_with_clf = keras.models.load_model(nameFile + '_encoder_clf_model_with_clf.h5', compile=False)
      self.model_without_clf = keras.models.load_model(nameFile+'_encoder_clf_model_without_clf.h5', compile=False)




 
   
```

# **Autoencoder convlocional**

```python
# Define the convolutional autoencoder model
input_shape = (252, 1)  # Each sample has 100 features, represented as a column vector
model = keras.Sequential([
    layers.Input(shape=input_shape),
    layers.Conv1D(16, 3, activation="relu", padding="same"),
    layers.MaxPooling1D(2, padding="same"),
    layers.Conv1D(8, 3, activation="relu", padding="same"),
    layers.MaxPooling1D(2, padding="same"),
    layers.Conv1D(8, 3, activation="relu", padding="same"),
    layers.MaxPooling1D(2, padding="same"),
    layers.Conv1D(8, 3, activation="relu", padding="same"),
    layers.MaxPooling1D(2, padding="same"),
    layers.Conv1D(8, 3, activation="relu", padding="same"),
    layers.UpSampling1D(2),
    layers.Conv1D(8, 3, activation="relu", padding="same"),
    layers.UpSampling1D(2),
    layers.Conv1D(8, 3, activation="relu", padding="same"),
    layers.UpSampling1D(2),
    layers.Conv1D(16, 3, activation="relu"),
    layers.UpSampling1D(2),
    layers.Conv1D(1, 3, activation="sigmoid", padding="same"),
])

# Compile the model
model.compile(optimizer="adam", loss="binary_crossentropy")

# Train the autoencoder
model.fit(x_train_norm, x_train_norm, epochs=10, batch_size=128, validation_data=(x_test_norm, x_test_norm))

# Use the trained autoencoder to reconstruct the test data
reconstructed_data = model.predict(x_test_norm)

```
