```python
import pydicom
import pydicom.data
import shutil
import os
```

```python
def anonymize_dicom_file(file_path, output_path):
    try:
        ds = pydicom.dcmread(file_path)
        print(f"\nAnonimizando archivo: {file_path}")

        # Anonimizar parametros
        ds.PatientName = ''  # Anonimizar el nombre del paciente
        ds.PatientID = ''  # Anonimizar el ID del paciente
        
        ds.ReferencedStudySequence.clear()
        ds.ReferencedPerformedProcedureStepSequence.clear()

        # Guardar el archivo en la misma carpeta
        ds.save_as(file_path)
        print(f"Archivo anonimizado guardado en: {file_path}")

    except Exception as e:
        print(f"No se pudo leer {file_path} como un archivo DICOM. Error: {str(e)}")
```

```python
def process_dcm_file(file_path, patient_folder):
    try:
        ds = pydicom.dcmread(file_path)
        print(f"Procesando archivo: {file_path}")

        # Verifica si es DTI o T1
        is_dti = 'DIFFUSION' in ds.ImageType and 'DTI' in ds.SeriesDescription
        is_t1 = 'T1' in ds.ImageType or 'T1' in ds.SeriesDescription

        if is_dti:
            print("Es imagen DTI")
        elif is_t1:
            print("Es imagen T1")

        if is_dti or is_t1:
            image_type = "DTI" if is_dti else "T1"
            
            base_name = os.path.basename(file_path)
            if not base_name.startswith(f"{image_type}_"):
                new_file_name = f"{image_type}_{base_name}"
            else:
                new_file_name = base_name
            
            new_file_path = os.path.join(patient_folder, new_file_name)

            # Renombrar el archivo
            shutil.move(file_path, new_file_path)
            print(f"Archivo renombrado a: {new_file_path}")
            
            # Llamar a la función de anonimización
            anonymize_dicom_file(new_file_path)
        else:
            # Si no es DTI ni T1, se elimina el archivo
            os.remove(file_path)
            print("Archivo no era ni DTI ni T1 y fue eliminado")

    except Exception as e:
        print(f"Error al procesar {file_path}: {str(e)}")
        # Si ocurre un error, se elimina el archivo
        if os.path.exists(file_path):
            os.remove(file_path)
            print("Archivo eliminado debido a un error durante el procesamiento")
```

```python
def check_files(folder_paciente):
    print("\n\n\n--------------------------------")
    print("Procesando paciente:", os.path.basename(folder_paciente))
    dicom_folder = os.path.join(folder_paciente, "Cerebro", "DICOM")

    if not os.path.exists(dicom_folder):
        print(f"La carpeta {dicom_folder} no existe.")
        return

    # Obtener todos los archivos en la carpeta DICOM, independientemente de la extensión
    all_files = [os.path.join(dicom_folder, f) for f in os.listdir(dicom_folder)]

    if not all_files:
        print("No hay archivos en la carpeta DICOM.")
        return

    print(f"Se encontraron {len(all_files)} archivos en la carpeta DICOM.")
    print(f"Procesando {len(all_files)} archivos.")

    for file_path in all_files:
        process_dcm_file(file_path, dicom_folder)
```

```python
# Listar todas las carpetas de pacientes y procesarlas
def process_patient_folders(main_folder):
    if not os.path.exists(main_folder):
        print(f"La carpeta {main_folder} no existe.")
        return

    patient_folders = [os.path.join(main_folder, f) for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]

    for folder in patient_folders:
        check_files(folder)
```

```python
def print_dicom_file_info(file_path):
    try:
        ds = pydicom.dcmread(file_path)

        # Imprimir la ruta del archivo
        print(f"\nArchivo: {file_path}")

        # Verificar si es DTI, T1, etc.
        is_dti = 'DIFFUSION' in ds.ImageType and 'DTI' in ds.SeriesDescription
        is_t1 = 'T1' in ds.ImageType or 'T1' in ds.SeriesDescription

        print("Características de la imagen:")
        if is_dti:
            print("- Tipo de imagen: DTI")
        elif is_t1:
            print("- Tipo de imagen: T1")
        else:
            print("- Tipo de imagen no identificado")

        # características del archivo DICOM
        print(f"- Descripción de la Serie: {ds.SeriesDescription}")
        print(f"- Tipo de Imagen: {ds.ImageType}")

        # header DICOM
        print("\nInformación del Header DICOM:")
        """
        for element in ds:
            print(f"{element.tag} ({element.name}): {element.value}")
        """
    except Exception as e:
        print(f"No se pudo leer {file_path} como un archivo DICOM. Error: {str(e)}")
```

```python
def check_files_in_folder(folder_path):
    print(f"\nVerificando archivos en la carpeta: {folder_path}")

    dicom_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    if not dicom_files:
        print("No se encontraron archivos en la carpeta.")
        return

    for file_path in dicom_files:
        print_dicom_file_info(file_path)
```

```python
def verify_dicom_folders(main_folder):
    if not os.path.exists(main_folder):
        print(f"La carpeta {main_folder} no existe.")
        return

    # Lista todas las carpetas de pacientes
    patient_folders = [os.path.join(main_folder, f) for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]

    for folder in patient_folders:
        patient_name = os.path.basename(folder)
        print(f"\n\n\n\n--- Procesando paciente: {patient_name} ---")

        # Misma ruta original
        dicom_folder = os.path.join(folder, "Cerebro", "DICOM")

        if os.path.exists(dicom_folder):
            check_files_in_folder(dicom_folder)
        else:
            print(f"La carpeta {dicom_folder} no existe para el paciente {patient_name}")
```

```python
# Iniciar la verificación
main_folder_path = "4_de_Abril"
verify_dicom_folders(main_folder_path)
```

```python
def rename_patient_folders(main_folder):
    if not os.path.exists(main_folder):
        print(f"La carpeta {main_folder} no existe.")
        return
    
    # Listar todas las carpetas dentro de la carpeta principal
    patient_folders = [os.path.join(main_folder, f) for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]
    
    # Contador para nombrar las carpetas de manera secuencial
    patient_counter = 1
    
    for folder in patient_folders:
        # Construir el nuevo nombre de la carpeta
        new_folder_name = f"paciente_{patient_counter}"
        new_folder_path = os.path.join(main_folder, new_folder_name)
        
        # Renombrar la carpeta
        os.rename(folder, new_folder_path)
        print(f"Carpeta {os.path.basename(folder)} renombrada a {new_folder_name}")
        
        # Incrementar el contador para el siguiente paciente
        patient_counter += 1

# Ruta de la carpeta principal donde están las carpetas de los pacientes
main_folder = "4_de_Abril"

# Llamada a la función para renombrar las carpetas
rename_patient_folders(main_folder)
```

```python
def add_dcm_extension(folder_path):
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if not file.endswith(".dcm"):
                os.rename(os.path.join(root, file), os.path.join(root, file + ".dcm"))
```

```python
def explore_dicom_fields(folder_path):
    for root, dirs, files in os.walk(folder_path):
        for dir in dirs:
            if dir == "Cerebro":
                dicom_folder_path = os.path.join(root, dir, "DICOM")
                if os.path.exists(dicom_folder_path):
                    add_dcm_extension(dicom_folder_path)  # Añade la extensión .dcm a las imágenes
                    for file in os.listdir(dicom_folder_path):
                        if file.endswith(".dcm"):
                            file_path = os.path.join(dicom_folder_path, file)
                            try:
                                ds = pydicom.dcmread(file_path)
                                print(f"Archivo: {file_path}")
                                study_description = ds.get("StudyDescription", "No disponible")
                                series_description = ds.get("SeriesDescription", "No disponible")
                                print(f"Study Description: {study_description}")
                                print(f"Series Description: {series_description}")
                                print("-" * 50)  # Imprime una línea divisoria
                            except Exception as e:
                                print(f"Error al leer {file_path}: {str(e)}")

# Ruta de la carpeta principal que contiene las carpetas de los pacientes
main_folder_path = "4_de_Abril"

# Llama a la función para explorar los campos
explore_dicom_fields(main_folder_path)
```
