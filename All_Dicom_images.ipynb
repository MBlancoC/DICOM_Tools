```python
def add_extension(file, extension):
    return file + extension
```

```python
import os
import pydicom

def is_dicom_file(file_path):
    try:
        with open(file_path, "rb") as f:
            dicom_file = pydicom.dcmread(f)
            return True
    except:
        return False
    
def check_files(folder_paciente):
    print("Paciente:", os.path.basename(folder_paciente))
    apellido = os.path.basename(folder_paciente).split("_")[0]
    dicom_folder = folder_paciente + "/Cerebro/DICOM"

    all_files = [os.path.join(dicom_folder, f) for f in os.listdir(dicom_folder)]
    print(len(all_files))

    dicom_files = [f for f in all_files if is_dicom_file(f)]
    print(dicom_files)

    for dicom_file in dicom_files:
        try:
            ds = pydicom.dcmread(dicom_file)
            print("Archivo DICOM:", os.path.basename(dicom_file))
            print(f"Descripción de la Serie: {ds.SeriesDescription}")
            print(f"Descripción de la Serie: {ds.ImageType}")

            #REVISA SI ES DTI
            if ds.ImageType == ['ORIGINAL', 'PRIMARY', 'DIFFUSION', 'NONE'] and ds.SeriesDescription == "DTI_15_NSA2":
                print("IMAGEN DTI")
                new_name = add_extension("DTI" + apellido + os.path.basename(dicom_file),".dcm")
                os.rename(dicom_file, new_name)
                print(f"Extensión agregada: {new_name}")

            #REVISA SI ES T1
            if ds.ImageType == ['ORIGINAL', 'PRIMARY', 'T1', 'NONE'] and ds.SeriesDescription == "SAG T1 3D":
                print("IMAGEN T1")
                new_name = add_extension("T1" + apellido + os.path.basename(dicom_file),".dcm")
                os.rename(dicom_file, new_name)
                print(f"Extensión agregada: {new_name}")
            # ELIMINA SI NO ES DTI O T1
            else:
                os.remove(dicom_file)
                print("Archivo eliminado")
            print("-" * 30)

        except Exception as e:
            #SI HAY ERROR ELIMINA EL ARCHIVO SI ES QUE AUN EXISTE EN LA CARPETA
            dicom_files = [f for f in all_files if is_dicom_file(f)]
            print(f"Error al procesar {dicom_file}: {str(e)}")
            if dicom_file in dicom_files:
                print("Existe")
                os.remove(dicom_file)
                print("Archivo eliminado")
            else:
                print("No existe")
            print("-" * 30)
    print(" ")
    print(" ")
    print(" ")
    print(" ")
    print(" ")
    print(" ")

#EXTRAE EL NOMBRE DE LAS CARPETAS DE LOS PACIENTES
def extraer_nombre(folder):
    all_files = [os.path.join(folder, f) for f in os.listdir(folder)]
    print(all_files[1:])
    return all_files[1:]
```

```python
patients = extraer_nombre("pacientes")
```

```python
for pat in patients:
    check_files(pat)
```
